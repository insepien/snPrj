```python
import numpy as np
from scipy import special
from math import factorial as fact
import matplotlib.pyplot as plt
import pandas as pd
from plotly.subplots import make_subplots
import plotly.graph_objects as go
from mpl_toolkits.axes_grid1 import make_axes_locatable

plt.rcParams['font.size'] = 15
plt.rcParams["axes.labelsize"] = 15
```

```python
def exponential_comp(t, thetaT, sigma, a):
    """gives the stuffs in the exponential"""
    tplus = (t - thetaT + a/2)  * sigma / np.sqrt(2)  
    tminus = (t - thetaT - a/2) * sigma / np.sqrt(2)
    return tplus, tminus

def d_erf(t, thetaT, sigma, a):
    """delta error function"""
    tplus, tminus = exponential_comp(t, thetaT, sigma, a)
    d_erf = 0.5 * (special.erf(tplus) - special.erf(tminus))
    return d_erf

def mu(x, y, thetaX, thetaY, sigma, thetaI0, a, thetabg):
    """image model for gaussian psf"""
    d_erfX = d_erf(x, thetaX, sigma, a)
    d_erfY = d_erf(y, thetaY, sigma, a)
    muK = thetaI0 * d_erfX * d_erfY + thetabg*a**2
    return muK
```

```python
sigma = 3
xmax = sigma*8
ymax = xmax
thetaX = (xmax-1)/2
thetaY = thetaX
a = 0.5
thetabg = 1
N = 500

x,y = np.arange(xmax), np.arange(ymax)
xs, ys = np.meshgrid(x,y)
#ys = np.flipud(ys)

m = mu(xs, ys, thetaX, thetaY, 1/sigma, N, a, thetabg)
plt.imshow(m,cmap="Greens")
plt.plot(thetaX,thetaY,"ro")
sE = np.sum(m)
sT = a**2*thetabg*xmax**2+N*a**2
sD = sT-sE
print(f"im: {sE:.3f}")
print(f"true: {sT:.3f}")
print(f"diff: {-sD:.3f}")
plt.colorbar();
```

```python
plt.rcParams["font.size"] = 15
a = np.arange(0.2,2,0.05)
m = [np.sum(mu(xs, ys, thetaX, thetaY, 1/sigma, N, a[i], thetabg)) for i in range(len(a))]
mT = a**2*thetabg*xmax**2+N*a*a
plt.plot(a,m,"b", label="Calculated")
plt.plot(a,mT,"g", label="Theoretical")
plt.xlabel("Pixel size a")
plt.ylabel("Photon counts N")
plt.legend()
```
